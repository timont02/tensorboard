```python
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import xml.etree.ElementTree as ET

directory = 'C:/Users/JannesPC/Documents/AiProjects/test'
print(directory)

for file in os.listdir(directory):
    if file.endswith('.jpg'):
        imgFilename = directory + '/' + file
        print(imgFilename)

        xmlFilename = imgFilename.rsplit('.')[0] + '.xml'

        xmlTree = ET.parse(xmlFilename)
        rootElement = xmlTree.getroot()

        for element in rootElement.findall("object"):
            xmin = 0
            ymin = 0
            xmax = 0
            ymax = 0

            for child in element:
                if child.tag  == "bndbox":
                    for child2 in child:
                        if child2.tag == "xmin":
                            xmin = int(child2.text)
                        elif child2.tag == "ymin":
                            ymin = int(child2.text)
                        elif child2.tag == "xmax":
                            xmax = int(child2.text)
                        elif child2.tag == "ymax":
                            ymax = int(child2.text)

                    x = xmin
                    y = ymin
                    dx = xmax - xmin
                    dy = ymax - ymin

                    bndbox = patches.Rectangle((x, y), dx, dy, linewidth=1, edgecolor='r', facecolor='none')


                    img = Image.open(imgFilename)
                    # Create figure and axes
                    fig, ax = plt.subplots()

                    # Display the image
                    ax.imshow(img)

                    ax.add_patch(bndbox)
                    plt.show()
```
